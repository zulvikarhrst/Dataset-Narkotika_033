#Library

```python
import requests
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

#Load

```python
url = 'https://github.com/zulvikarhrst/Dataset-Narkotika_033/raw/main/Overview/Overview.xlsx'
response = requests.get(url)
with open('Overview.xlsx', 'wb') as f:
    f.write(response.content)
```

```python
df = pd.read_excel("Overview.xlsx")
df
```

#Pre Processing

```python
def preprocess_text(text):

    if pd.isnull(text):
        return ''
    text = text.lower()  # Ubah menjadi huruf kecil
    text = re.sub(r'[^a-z\s]', '', text)  # Hapus karakter khusus dan angka
    text = re.sub(r'\s+', ' ', text).strip()  # Hapus spasi ekstra
    return text

df['Barang Bukti Preprocessed'] = df['Barang Bukti'].apply(preprocess_text)
df['Amar Putusan Preprocessed'] = df['Amar Putusan '].apply(preprocess_text)

df[['Barang Bukti', 'Barang Bukti Preprocessed', 'Amar Putusan ', 'Amar Putusan Preprocessed']]
```

```python
# Menggabungkan kolom 'Barang Bukti' dan 'Amar Putusan' menjadi satu dokumen
documents = df['Barang Bukti'] + ' ' + df['Amar Putusan ']

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Melatih model K-Nearest Neighbors menggunakan metrik cosine similarity
from sklearn.neighbors import NearestNeighbors
knn_model = NearestNeighbors(n_neighbors=5, metric='cosine')
knn_model.fit(tfidf_matrix)

# Fungsi untuk mencari dokumen yang mirip dengan query tertentu
def search_documents(query):
    query_vec = vectorizer.transform([query])
    distances, indices = knn_model.kneighbors(query_vec)
    for i, index in enumerate(indices[0]):
        print(f"Rank {i + 1}:")
        print(f"Document Index: {index}")
        print(f"Similarity Score: {1 - distances[0][i]}")
        print(f"Barang Bukti: {df['Barang Bukti'].iloc[index]}")
        print(f"Amar Putusan: {df['Amar Putusan '].iloc[index]}\n")

# Contoh pencarian
search_query = "narkotika dan psikotropika"
search_documents(search_query)
```
